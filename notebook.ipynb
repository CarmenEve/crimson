```python
import json
```

```python
with open('./entities.json') as f:
  data = json.load(f)
```

```python
appositives = ['writer ', 
               'freshman ', 'sophomore ', 'junior ', 'senior ', 
               'hitter ', 'attackman ', 'veterans ',
               'poet ', 'cinematographer ',
               'councilor ', 'councillor' , 'faculty dean ', 'dean ', 'resident dean ', 'chair ']
```

```python
import csv

facebook_names = []

with open("facebook.csv") as f:
    f_data_reader = csv.DictReader(f, delimiter=",")
    for row in f_data_reader:
        facebook_names.append(row["First Name"] + " " + row["Last Name"])
```

```python
# fixing misnames
for entity in data:
    for i in range(len(entity["named_entities"])):
        name = entity["named_entities"][i]
        # some names begin with appositives
        for a in appositives: 
            if name.lower().find(a) == 0:
                entity["named_entities"][i] = name[len(a):]
                break
        # some names include the ’ or end with ’s
        if len(name) >= 2 and (name[-2:] == " ’" or name[-2:] == "’s" or name[-2:] == "'s"):
            entity["named_entities"][i] = name[:-2]
        # some names include the ’(year)
        elif len(name) >= 3 and name[-3] == "’" and name[-2:].isnumeric():
            entity["named_entities"][i] = name[:-3]

# remove repeat names
for entity in data:
    entity["named_entities"] = list(dict.fromkeys(entity["named_entities"]))

# removing non-names
for i in range(len(data)):
    new_named_entities = data[i]["named_entities"].copy()
    for name in data[i]["named_entities"]:
        name_list = name.split()
        # full names are at least 2 words
        if len(name_list) < 2:
            new_named_entities.remove(name)
        # remove non-facebook names
        elif name_list[0] + " " + name_list[-1] not in facebook_names:
            new_named_entities.remove(name)
    data[i]["named_entities"] = new_named_entities

# removing all entities with empty named_entities
new_data = data.copy()
for entity in data:
    if not entity["named_entities"]:
        new_data.remove(entity)
data = new_data
```

```python
from random import randrange

i = randrange(len(data))
print(data[i]["url"])
print(data[i]["named_entities"])
```

```python
# testing
dic = {}
for entity in data:
    for name in entity["named_entities"]:
        if name in dic:
            dic[name] += 1
        else: 
            dic[name] = 1

from collections import Counter 

k = Counter(dic) 
high = k.most_common(20)  
for i in high: 
    print(i[0], ":" ,i[1]) 
```

```python
with open('cleaned_entities.json', 'w') as f:
    json.dump(data, f)
```
